**Pie Chart as a format**

Pros
1. Simplicity: This pie chartis good for presenting a simple distribution of a Survey Results as a whole, making them easy for people to quickly grasp.
2. Comparison of proportions: The each segment compares to the total, this pie chart does a good job of providing that visual.
3. Clear Segments: The pie chart breaks down the responses into clear categories, which helps show the relative proportions of each sentiment.
4. The exact percentage values directly to the pie chart slices. This allows for immediate interpretation without having to reference the legend.

Cons
1. Limited Detail: Pie charts work best with a small number of categories. While this chart has a manageable number of categories (5), it's still somewhat hard to compare values accurately because the differences between the slices (such as between "OK" and "Kind of interested") are subtle.
2. Inaccuracy of Estimation: It's hard for viewers to estimate precise percentages from the chart, especially with certain slices being very close in size (like "Excited" at 19% and "Kind of interested" at 25%).
3. Ambiguity in Categories: The labels themselves could be more precise. For example, "Kind of interested" is a subjective label that could mean different things to different people, and "Not great" might be too vague.

**Color Scheme**
1. The color choices are varied, which can help distinguish between the categories. The use of different colors for each slice allows viewers to visually identify each sentiment response easily.
2. Accessibility Concerns: The colors might not be distinguishable for individuals with color blindness. For example, dark blue and purple can appear similar to some viewers, making it harder to differentiate the "Bored" and "Kind of interested" slices.
3. Too Many Different Colors: While the colors are varied, there may be too many hues, which could make the chart look busy or cluttered. Some of the colors (like light blue and light green) may not stand out enough, especially in a print format.

**Legend**
1. The legend clearly matches each sentiment category with the corresponding color, making it easy for viewers to identify which segment represents which sentiment.

**Title**
1. "Survey Results" is clear, but it lacks context. It tells the audience that the data is from a survey but doesn't explain what the survey is about.
2. Lack of Specificity: The title doesn’t give viewers a sense of what exactly was surveyed, which makes it harder to understand the context of the data. Without context, the pie chart could feel generic and ambiguous.

**How I Would Draw It Differently:**
1. Consider a Bar Chart: If the goal is to compare the sizes of each sentiment more easily, a bar chart would be a better choice. Bar charts allow for more precise comparisons of the percentage differences between categories. A dot plot could also work, providing a more accurate way to visualize the proportions.
2. Improve Color Contrast: I’d choose a more cohesive and contrasting color palette, making sure the colors are distinct and easily readable for those with color vision deficiencies. For instance, I might use distinct shades like dark green, orange, blue, purple, and yellow, which are more easily distinguishable by a wider range of people.
3. Clarify Title: I’d refine the title to something more specific, such as "Survey Results: How People Feel About the Upcoming Event," which gives the reader immediate context about the data.
4. Consider Adding Context in the Chart: If using a pie chart, including the data source or a brief description of the survey within the chart could also help provide clarity.

8. What are the mean, median, and mode customer age in years?  (Rounding down to the next lower age.)
Are there any outliers?  (Customers with very large or very small ages, compared with the other ages?)

```python
# Step 1: Calculate age (current date - birthdate), and round down to next lower age
current_date = pd.to_datetime('today')
df_bank_uniquefix['Age'] = (current_date - df_bank_uniquefix['BirthDate']).dt.days // 365  # Using integer division for rounding down
```

```python
# Step 2: Calculate mean, median, and mode of the ages
mean_age = round(df_bank_uniquefix['Age'].mean())
median_age = round(df_bank_uniquefix['Age'].median())
mode_age = round(df_bank_uniquefix['Age'].mode())[0]  # mode() returns a Series, get the first mode

# Step 3: Detect outliers using IQR (Interquartile Range) method
Q1 = df_bank_uniquefix['Age'].quantile(0.25)
Q3 = df_bank_uniquefix['Age'].quantile(0.75)
IQR = Q3 - Q1
print(f"Q1 percentile: {Q1}")
print(f"Q3 percentile: {Q3}")
print(f"IQR: {IQR}")

# Step 4: Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("\nLower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")

# Step5: Filter customers with outliers
outliers = df_bank_uniquefix[(df_bank_uniquefix['Age'] < lower_bound) | (df_bank_uniquefix['Age'] > upper_bound)]

# Output results
print("\nMean Age: {mean_age}")
print(f"Median Age: {median_age}")
print(f"Mode Age: {mode_age}")
print("\nOutliers (customers with unusual ages):")
print(outliers[['CustomerID', 'Age']])
```

```python
# Check Maxm & Minm Ages
maxm_age = round(df_bank_uniquefix['Age'].max())
minm_age = round(df_bank_uniquefix['Age'].min())
print(f"Maximum Age: {maxm_age}")
print(f"Minimum Age: {minm_age}")
```

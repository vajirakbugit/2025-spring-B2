# 2. Weekly graph question

Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

**ANSWER**
# Advantages of an Area Plot
1. Visual Emphasis on Cumulative Trends :
1.1. If the goal is to show how nitrate and phosphate levels contribute to the overall "total" nutrient level in the water sample, an area plot can visually emphasize this cumulative effect.
1.2. The stacked nature of the plot makes it easy to see the combined impact of both nutrients.

2. Comparison of Relative Contributions :
2.1. The relative proportions of nitrate and phosphate levels are more apparent in an area plot because the height of each segment corresponds to its contribution.

3. Aesthetic Appeal :
3.1. Area plots can be visually appealing and engaging, especially when color gradients or transparency are used to distinguish between the two nutrients.

# Disadvantages of an Area Plot
1. Difficulty in Reading Exact Values :
1.1. It can be harder to discern the exact values of nitrate and phosphate levels, especially for the nutrient plotted on top (e.g., phosphate), because its baseline is not fixed at zero but instead depends on the nitrate level below it.

2. Potential for Misinterpretation :
2.1. Viewers might mistakenly interpret the total height of the plot as representing a single variable rather than the sum of two separate 
variables (nitrate and phosphate).

3. Limited Usefulness for Non-Cumulative Data :
3.1 If nitrate and phosphate levels are independent and do not naturally combine into a meaningful total, an area plot may not be appropriate.

# Advantages of Two Separate Line Graphs
1. Clarity of Individual Trends :
1.1. Two separate line graphs allow viewers to easily track the trends of nitrate and phosphate levels independently without interference from overlapping or stacking.

2. Ease of Reading Exact Values :
2.1. Each line graph has its own axis, making it straightforward to read exact values for each nutrient at any given point.

3. Flexibility for Independent Variables :
3.1. If nitrate and phosphate levels are unrelated or measured on different scales, two separate line graphs avoid forcing them into a combined representation.

# Disadvantages of Two Separate Line Graphs
1. Lack of Visual Comparison :
1.1. It is harder to visually compare the relative contributions of nitrate and phosphate levels at a glance compared to an area plot.

2. Increased Space Usage :
2.1. Two separate graphs require more space, which can make the presentation feel cluttered or less cohesive.

3. Reduced Emphasis on Combined Effect :
3.1. If the goal is to highlight the total nutrient level, two separate line graphs do not emphasize the cumulative effect as effectively as an area plot.

**How to Adjust the Graphs for Better Presentation**
# Improving an Area Plot
1. Use Transparency : Apply transparency (alpha blending) to the filled areas so that overlapping sections remain visible and do not obscure underlying data.
2. Add Clear Labels : Label the axes clearly and include a legend to identify which nutrient corresponds to each color.
3. Highlight Key Points : Add markers (e.g., dots or circles) at key points on the plot to make it easier to read exact values.
4. Limit Stacking : If the cumulative effect is not relevant, consider using a non-stacked area plot where each nutrient is plotted independently with its baseline at zero.
5. Include a Total Line : Add a dashed or bold line to represent the total nutrient level if the cumulative effect is important.

# Improving Two Separate Line Graphs
1. Align Axes : Ensure both graphs use the same scale on the y-axis (if possible) to facilitate comparison between nitrate and phosphate levels.
2. Use Consistent Colors : Assign consistent colors to nitrate and phosphate across both graphs to reinforce their association.
3. Combine into a Single Chart : Overlay the two lines on the same graph with distinct line styles (e.g., solid for nitrate, dashed for phosphate) and a shared y-axis.
4. Add Annotations : Highlight key trends or anomalies with annotations or arrows to draw attention to important features.
5. Provide Context : Include a title or caption explaining the significance of the data (e.g., whether high nitrate or phosphate levels indicate pollution).

# Use an area plot if:
1. If there is a need to emphasize the cumulative effect of nitrate and phosphate levels.
2. The relative contributions of each nutrient are important.

# Use two separate line graphs if:
1. If there is requirement to focus on the individual trends of nitrate and phosphate levels.
2. The nutrients are independent and do not naturally combine into a meaningful total.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
df = pd.DataFrame({'ser1': time_series, 'ser2': time_series_2}, index = datearray)
df.plot.area()
```

```python
# As it says to utilize the last semester data set for the Week6
import math
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

red_wine = pd.read_csv("https://github.com/EpistasisLab/pmlb/raw/refs/heads/master/datasets/wine_quality_red/wine_quality_red.tsv.gz", sep="\t")
red_wine.info()
```

**Find correlations between pairs of variables.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
correlation_matrix = red_wine.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix Heatmap", fontsize=16)
plt.tight_layout()
plt.show()
```

**Draw scatterplots, especially when the correlation is large**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Identify pairs with high correlation (absolute value > 0.5)
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.5:
            high_corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))

# Plot scatterplots for high-correlation pairs
plt.figure(figsize=(15, 10))
for idx, (x, y) in enumerate(high_corr_pairs, 1):
    plt.subplot(2, 3, idx)  # Adjust grid size based on number of pairs
    sns.scatterplot(data=red_wine, x=x, y=y, alpha=0.7)
    plt.title(f"Scatterplot: {x} vs {y}", fontsize=12)
    plt.xlabel(x, fontsize=10)
    plt.ylabel(y, fontsize=10)

plt.tight_layout()
plt.show()
```
